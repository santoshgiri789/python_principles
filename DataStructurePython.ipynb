# Example: (List+For+If-else+Operators) - Given a list of numbers, calculate the sum of all the numbers as well as the product of all the numbers, but do not include multiples of 5 in the computation for either result.

````python
numbers = [1,2,3,4,5,6,7,8] 
sum = 0
product = 1
for i in numbers:
    if(i % 5 != 0):
        sum = sum + i
        product = product * i
    else:
        continue

print('Sum of all elements:', sum)
print('Product of all elements:', product)
````

````python
### Example You are given a list of customer names, but the list contains duplicates. Return a list of the names without duplicates.
customer_names = ['Alisha', 'Bob', 'Charlie', 'Alisha', 'Eshan', 'Bob', 'Frank', 'Greeshma', 'Charlie', 'Hannah']
unique_customer_names = []
for name in customer_names:
    if name not in unique_customer_names:
        unique_customer_names += [name]
print(unique_customer_names)
````

### Example
# The following Python code is intended to create a new list where even numbers are doubled but odd numbers remain unchanged.
# However, the code results in an error. Identify the error and rewrite the list comprehension correctly.


````python
numbers= [1,3,2,4,7]
new_numbers = []
for intnum in numbers:
    if intnum % 2 == 0:
        new_numbers += [intnum*2]
    else:
        new_numbers += [intnum]

print(new_numbers)
````
