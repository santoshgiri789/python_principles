```python
### Write a function to calculate the simple interest and final amount for the given principal, time duration and rate of interest
```

```python
def calculateSimpleInterest(principal, time_duration, rate_of_interest): 
    simple_interest = (principal* rate_of_interest * time_duration)/100
    print("Simple Interest:",simple_interest)
    return  simple_interest

def calculateTotalAmount(principal, time_duration, rate_of_interest):
    si = calculateSimpleInterest(principal, time_duration, rate_of_interest)
    amount = principal + si
    return amount;

amt = calculateTotalAmount(10000, 5, 12)
print("Total amount:",amt)



```

```python
# ### Write a python function to calculate the actual price of the product which is 
# summation of amount (user amount) and gst, by default is 5%, but can have other values as well
```

```python
def calculateActualPrice(amount, gst=5):
    totalAmount = amount + (amount * gst/100)
    return totalAmount

calculateActualPrice(1000,20.1)
```

```python
# Even or Odd
def evenOrOdd1(number):
    if number%2 == 0:
        return True
    else:
        return False
def evenOrOdd2(number):
    return number%2 == 0
    
print(evenOrOdd1(0))
print(evenOrOdd2(5))
```

```python
### Write a python function to assign a age group to a user input age value basis below criteria
# - < 18 - Kid,  >=18 and <=30 - Young,  >30 and <=50 - Middle Age,  >50 - Old
```

```python
age = int(input("Enter the age: "))

def assigneAgeGroup(ageNumber):
    
    if ageNumber < 18:
        response =  "Kid"
    elif ageNumber >= 18 and ageNumber <= 30:
        response = "Young"
    elif ageNumber > 30 and ageNumber <= 50:
        response = "Middle Age"
    elif ageNumber > 50:
        response = "Old"
    return response

assigneAgeGroup(age)
```

```python
## Lambda - Anonymous Functions
# Small one line function
# No function name (not mandatory)
# Used for short operations 
```

```python
add = lambda a, b: a + b
add(10,20)
```

```python
is_even = lambda x : x%2 == 0
is_even(10)
```

```python
ls=['apple','guava','pienapple','litchi'] # map is python function which takes each element of list as input and apply a function on each value.
func=lambda x:x.upper()
list(map(func, ls))
```
