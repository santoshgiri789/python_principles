## Data Structure - non primitive
## Start List

```python
shopping_cart = ["Bagpack", "Charger"]
print("Before:",shopping_cart)
shopping_cart.extend(["Coconut"])
print("After:",shopping_cart)
```

```python
print(shopping_cart[2])
print(shopping_cart[-1])
print(shopping_cart[0])
print(shopping_cart[-2])
```

```python
## Removing elements from List
# Method1
print("Before remove:",shopping_cart)
shopping_cart.remove("Coconut")
print("After remove:",shopping_cart)
```

```python
# Method2
print("Before remove:",shopping_cart)
del shopping_cart[2]
print("After remove:",shopping_cart)
```

```python
# Method3
print("Before pop:",shopping_cart)
shopping_cart.pop(1)
print("After pop:",shopping_cart)
```

## sorting elements - List

```python
print("Before sorting:",shopping_cart)
shopping_cart.sort(reverse=True)
print("After sorting:",shopping_cart)
```

# Set Starts - Unique elements, donot support indexing

```python
colors = {'Red','Pink','Orange','Black','Orange','Blue'}
print(colors)
```

```python
for color in colors:
    print(color)
```

```python
saleTeam = {'Santosh','Ramesh','Gopal','Srinath','Rahul','Sambhu','Parag'}
print('saleTeam', saleTeam)
marketingTeam = {'Piyush','Mahesh','Parag'}
print('marketingTeam', marketingTeam)


```

```python
saleTeam = saleTeam.intersection(marketingTeam)
print('Intersection Team:', saleTeam)
saleTeam = saleTeam.union(marketingTeam)
print('Union Team:', saleTeam)
```

# Dictionary Start - Key, Value

```python
usa = {'CA': "California", "Al": "Alabama", "KT": "Kentucky"}
print("Before:",usa)
print(usa.get("CA"))
print(usa.get("NJ"))
print(usa.get("NY", "New York not availbale"))
del usa['CA']
print("After:",usa)
```
