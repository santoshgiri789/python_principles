# Write a func!
# Given an integer x,
# return x with its digits reversed

# Example 1:
x = 123
# ●	Output: 321

# Example 2:
x = -123
# ●	Output: -321

```python
def rev(x):
  if x > 0 :
    print(int(str(x)[::-1]))
  else :
    print(int(str(x)[1:][::-1])*-1)

rev(-123)
```

# Dictionaries and Functions: Write a function update_scores
# that takes a dictionary containing player names and their scores,
#     and a list of tuples where each tuple contains a
#     player's name and points to add to their score.

player_scores = {'Alice': 10,   'Bob': 8,  'Carol': 15}
updates       = [('Alice', 5), ('Bob', 2), ('Dave', 7)]

# O/P : {'Alice': 15, 'Bob': 10, 'Carol': 15, 'Dave': 7}

```python
def update_scores(player_scores, updates):
    for update in updates:
        name  = update[0]
        newScore = update[1]
        if name in player_scores:
            print(name)
            score = player_scores[name]
            score = score + newScore
            player_scores.update({name: score})
        else:
            player_scores[name] = newScore
    print('player_scores', player_scores)
player_scores = {'Alice': 10,   'Bob': 8,  'Carol': 15}
updates       = [('Alice', 5), ('Bob', 2), ('Dave', 7)]
update_scores(player_scores, updates)
```
