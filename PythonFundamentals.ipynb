```python
# Objective: ask the user to enter the basic Pay ,

# if basicPay <=5000 , then calculate HRA as 16% of basicPay , and DA as 14% of basicPay

# if basicPay >5000, then Caclulate HRA as 12% of basicPay and DA as 10% of basicPay

#  find gross Pay i.e. sum of basicPay + HRA + DA
```

```python
basicPay = int (input('enter basic pay 💴 '))

if basicPay > 5000 :

  HRA= .12 * basicPay
  DA = .1  * basicPay

else:

  HRA = .16 * basicPay
  DA  = .14 * basicPay

# let us find the grossPay

grossPay = basicPay + HRA + DA

# let us use string formatting using F-String to print the salary details

salarySummary =f'''
-------------------------Salary Slip -----------------
BasicPay : {basicPay}
HRA     :  {HRA}
DA      :  {DA}
-------------------
Gross :    {grossPay}
'''

print( salarySummary )
```

```python
Step 1 - ask the user to enter a plain text to encrypt say 'apple' 🍎

Step 2 - Reverse the given Text i.e.

'apple' becomes 'elppa'

Step 3 - replace Vowels with 'digits' as per the details below:

'a' -> '0'
'e' -> '1'
'i' -> '2'
'o' -> '3'
'u' -> '4'

i.e. 'elppa' becomes '1lpp0'

Step 4: add 'aca' at the end of text

i.e.   '1lpp0' become '1lpp0aca'

print('the encrypted text i.e. 1lpp0aca') END
```

```python
stringInput = input('Enter the pain text:')
reverseInput = stringInput[::-1]
print(reverseInput)
vowels = ['a','e','o','i', 'u']
for x in reverseInput:
    if x in vowels:
        index = vowels.index(x)
        reverseInput = reverseInput.replace(x, str(index))
print(reverseInput+'aca')
        
```

```python
# You are Provided with a Guest List for an Upcoming AI EVENT
# You are asked to Print the Invitations in the below Foramt :
# Dear Mr/Ms      Kumar.R
# Dear Mr/Ms      Chauhan.A
# Dear Mr/Ms      Gupta.N
# Dear Mr/Ms      Mehta.R
# Dear Mr/Ms      Yadav.V
```

```python
# 1. Observe the Input Data ( List Containing Strings)
#  1.1  Each String has Full Name Separated By 'Space'

# 2. Split Each String using Space to form a List ( using .split() function)

# 3. Take the Last Element of Each List as it will give you LastName

# 4. Take the First Character of Each Element as it will give you First name first Character.
names = ["Rahul Kumar" , "Amit Singh Chauhan" , "Neha Gupta" ,"Rohit Kumar Mehta" ,"Vikas Yadav"]
```

```python
for name in names:
    #print(name)
    nameList = name.split(' ')
    #print(nameList)
    firstName = nameList[0];
    lastName = nameList[1];
    newName = lastName + '.' + firstName[0]
    print('Dear Mr/Ms ' + newName)
```

```python
# lets try using SET Data Structure,   this time you are asked to create a new set

# say uniqueTypes= set()

# write logic to add all Types into uniqueTypes Set and display the Same.

# create an empty set first

uniqueTypes= set()
# write a loop that fetches each element
for p in pokemons:
  uniqueTypes.add(p.split(',')[1].strip())
print(uniqueTypes)
print(len(uniqueTypes))
```

```python
### You  are asked to Create a Dictionary Containing Pokemon Statistics where
 #Key : pokemonType
 #value : Count of Pokemon Type

#for eg.

#{'fire':41 , 'water':58 , 'ghost':23 ......}
poke_stats= {}
for p in pokemons :
    ptype = p.split(',')[-1].strip()
    #print(poke_stats.keys)
    if ptype in poke_stats.keys():
        poke_stats[ptype] = poke_stats[ptype] + 1
    else:
        poke_stats[ ptype ] = 1
print(poke_stats)
```
