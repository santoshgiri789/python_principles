### Question 1 A bank application is designed to generate a formatted customer account statement. The statement should display the account holder’s name, account number, and account balance in a structured format. However, the current program contains several syntax errors and formatting mistakes. The statement should display all the details in the correct order and on separate lines exactly as shown in the sample output. You must print the entire statement using only one `print()` statement. Your task is to identify and fix all the errors so that the program prints the bank account statement exactly as shown in the expected output. Ensure the output format matches the sample output exactly. Before printing the statement, the program must verify whether the account number is a valid 12-digit number. - If the account number is not a 12-digit number, the program should print: `Invalid Account Number` - If the account number is valid, the program should print the formatted bank account 


````python
account_number = input('Enter the acoount number:')
output = ''
if len(account_number) > 12:
    output = 'Invalid Account Number: ' + account_number
else:
    acc_holder_name = input('Enter Account Holder Name:')
    acc_balance = int(input('Enter Account balance:'))
    output = f'''
    Bank Account Statement
    Account Holder: {acc_holder_name}
    Account Number: {account_number}
    Balance:        {acc_balance}
    Thank you for banking with us!
    '''
print(output)
    
    
````

### Question 4 Sentiment analysis is a technique used in text processing to determine whether a sentence expresses a positive, negative, or neutral opinion. In this problem, you are given a sentence as input. Your task is to analyse the sentence and classify its sentiment based on the presence of certain keywords. The sentiment should be determined using the following rules: - If the sentence contains any positive words such as `'good'`, `'happy'`, or `'excellent'`, the sentiment should be classified as `'Positive'` - If the sentence contains any negative words such as `'bad'`, `'sad'`, or `'terrible'`, the sentiment should be classified as `'Negative'` - If the sentence contains both positive and negative words, the sentiment should be classified as `'Neutral'` - If the sentence does not contain any of these positive or negative keywords, the sentiment should be classified as `'Neutral'`

````python
sentence = input('Enter the feedback:')
positive = ['good', 'happy', 'excellent']
negative = ['bad', 'sad', 'terrible']
positiveFeed = False
for i in (positive):
    if i in sentence:
        positiveFeed = True
        break
    else:
        continue
print('Feed1:',positiveFeed)
negativeFeed = False
for j in negative:
    if j in sentence:
        negativeFeed = True
        break
    else:
        continue

print('Feed2:',negativeFeed)
if (positiveFeed and negativeFeed):
    print('Neutral')
elif positiveFeed:
    print('Positive')
elif negativeFeed:
    print('Negative')
else:
    print('Neutral')
````

### Question 3 A website checks the strength of user passwords based on certain rules. The password must satisfy the following conditions: - Minimum length of 8 characters - Must contain at least one digit - Must contain at least one uppercase letter Write Python code to check whether a given password satisfies these conditions. If all conditions are satisfied, print `'Strong Password'`. Otherwise print `'Weak Password'`. **Input Format** A single string representing the password **Output Format** `Strong Password` or `Weak Password`
**Sample Test Case 1 - Input** `'Hello123'` **Sample Test Case 1 - Output** `Strong Password` **Sample Test Case 2 - Input** `'hello12'` **Sample Test Case 2 - Output** `Weak Password`

````python
password_str = input('Enter the password:')
lengthFlag = False
caseFlag = False
digitFlag = False
if len(password_str) > 8:
    lengthFlag = True
if any(char.isupper() for char in password_str):
    caseFlag = True
if any(char.isdigit() for char in password_str):
    digitFlag = True

if lengthFlag and caseFlag and digitFlag:
    print("'Strong Password'")
else:
    print("'Weak Password'")
````

### Question 6 Alarm Clock Do you wake up at the same time every morning? No, right? If it is a normal working day where you're swamped with work and meetings starting early morning, you try to wake up as soon as possible. If it is a weekend or a vacation day, you mostly want to give yourself some extra rest, right? Now, you might set alarms for this manually periodically depending on what day it is. Instead of manually setting up an alarm, what if you try to write a code that automates the time the alarm goes off. If there was some way for the alarm to know the kind of day it is, it would automatically ring according to the preferences set by you. Let's attempt to do this.


# So the problem statement is, you're trying to automate your alarm clock by writing a function for it. You're given a day of the week encoded as 1 = Mon, 2 = Tue, ..., 6 = Sat, 7 = Sun, and whether you are on vacation as a boolean value (a boolean object is either True or False. Google "booleans Python" to get a better understanding). Based on the day and whether you're on vacation, write a function that returns a time in form of a string indicating when the alarm clock should ring. When not on a vacation, on weekdays, the alarm should ring at "7:00" and on the weekends (Saturday and Sunday) it should ring at "10:00". While on a vacation, it should ring at "10:00" on weekdays. On vacation, it should not ring on weekends, i.e., it should return "off".

````python
day_of_week = int(input())
is_on_vacation = input()
if day_of_week in [1,2,3,4,5] and is_on_vacation=='False':
    time = '7:00'
elif day_of_week in [6,7] and is_on_vacation=='True':
    time = 'off'
else:
    time = '10:00'
print(time)
````
