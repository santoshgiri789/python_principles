```python
### Example Write `if-elif-else` statements for the following set of rules on a variable called `attendance` 
## - If attendance is perfect ($30$ days), 
## label the employee as `outstanding` 
## - If attendance is between $25$ and $29$ days (inclusive), 
## label the employee as `satisfactory` 
## - If attendance is between $20$ and $24$ days (inclusive), 
## label the employee as `fair` 
## - If attendance is below $20$ days, 
## label the employee as `poor`
```

```python
attendance = 10
if attendance == 30:
    print('outstanding')
elif attendance >= 25 and attendance <= 29:
    print('satisfactory')
elif attendance >= 20 and attendance <= 24:
    print('fair')
else : 
    print('poor')
```

# The Break Statement

```python
for num in range(10):
    print(num)
    if num == 5:
        print('Number 5 found')
        break
    else:
        print('continue')
        continue
```

### Nested For loops

```python
for i in range (1,4):
    for j in range(1,4):
        print(f'{i} * {j} = {i*j}')
```

### The While loop

```python
is_running = True
count = 0
while is_running:
    count += 1
    print('This is statement number', count)    
    if count >= 5:
        is_running = False
```
